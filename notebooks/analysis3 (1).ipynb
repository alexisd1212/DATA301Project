```python
def load_and_process(url_or_path_to_csv_file):

    # Method Chain 1 (Load data and deal with missing data)

    df1 = (
          pd.read_csv('./Downloads/application_record.csv.zip')
          .rename(columns={'CODE_GENDER': 'Gender', 'FLAG_OWN_CAR': 'own_car', 'CNT_CHILDREN': 'children', 'AMT_INCOME_TOTAL': 'income',
                          'NAME_INCOME_TYPE': 'Type of Employment', 'NAME_EDUCATION_TYPE': 'Level of Education', 'NAME_FAMILY_STATUS': 'relationship_status',
                          'NAME_HOUSING_TYPE': 'Housing'})
          .fillna("Inconlusive")
          .replace("F", "Female")
          .replace("M", "Male")
          .replace("Y", "Yes")
          .replace("N", "No")
        
    )

    # Method Chain 2 (Create new columns, drop others, and do processing)
    
    df2 = (
          df1.drop(['FLAG_OWN_REALTY', 'DAYS_BIRTH',
       'DAYS_EMPLOYED', 'FLAG_MOBIL', 'FLAG_WORK_PHONE', 'FLAG_PHONE',
       'FLAG_EMAIL'], axis=1)
    )

    return df2
```

```python
df3 = load_and_process('./Downloads/application_record.csv.zip')
```

```python
import seaborn as sns
sns.set_theme(style="whitegrid")

# Draw a nested barplot by education status and sex
g = sns.catplot(
    data=df3, kind="bar",
    x="own_car", y="income", hue="Level of Education",
    ci="sd", palette="dark", alpha=.6, height=6
)
g.despine(left=True)
g.set_axis_labels("", "Income")
g.legend.set_title("")
g.fig.suptitle("Relationship between Education level and income by car ownership")
g.fig.set_size_inches(20,5)
plt.show()
```

```python
#Analysis:
#-Car ownership has higher income in all educational levels 
#-If you own a car that means you are more likely to have the income to afford a car
#-Income correlates with higher education as expected
```

```python
import seaborn as sns
sns.set_theme(style="whitegrid")

# Draw a nested barplot by education status and sex
g = sns.catplot(
    data=df3, kind="bar",
    x="Level of Education", y="income", hue="Type of Employment",
    ci="sd", palette="dark", alpha=.6, height=6
)
g.despine(left=True)
g.set_axis_labels("", "Income")
g.legend.set_title("")
g.fig.suptitle("Relationship between Education level and income by employment type")
g.fig.set_size_inches(20,5)
plt.show()
```

```python
#Analysis

#The higher the educational level, the higher the income. Academic degree > Higher Education > Incomplete Higher > Secondary > Lower Secondary
#Commercial associate is the highest in all and pensioner is the lowest except lower secondary
```

```python
sns.relplot(
    data=df3, x="Level of Education", y="ID",
    col="Type of Employment",
    kind="scatter"
)
```

```python
#Analysis
#-Academic degree holders had the least amount of card ownership in all types of employment groups
#-The other levels of education ahd similar levels of card ownership
#-This is possibly because people who have the highest form of education have the most money and don't need to rely on credit cards as often
```

```python
plt.figure(figsize = (10, 10))
p = sns.violinplot(x=df3['ID'], y = df3['Type of Employment'], data = df3)
p.set_xlabel('ID', fontsize = 20)
```

```python
#Analysis
#-Students had the least amount of credit card ownership
#-possibly because many are underage and not mature enough yet
#-State servant, Pensioner, Commercial associaate and working class all exhibited similar levels of card ownership
```
